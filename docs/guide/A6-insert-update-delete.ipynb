# Insert, update & delete data: `db.insert, db.update & db.delete`

```python
import lamindb as ln
import pandas as pd

ln.nb.header()
```

## Insert

```python
biosample_id = ln.db.insert.biosample(name="B1")

biosample_id
```

Batch insert can be done by providing a DataFrame

```python
biosample_df = pd.DataFrame({"name": ["B2", "B3", "B4"]})

biosample_df
```

```python
ln.db.insert.from_df(biosample_df, "biosample")
```

Or by providing a list of kwargs

```python
ln.db.insert.from_list("disease", [dict(name="B5"), dict(name="B6"), dict(name="B7")])
```

## Update

```python
dobject = ln.db.select.dobject(name="iris").first()
dobject
```

Let's update the name "iris" to "iris_new".

```python
ln.db.update.dobject(dobject.id, name="iris_new")
```

Let's select the same object again to see if the name has changed.

```python
dobject = ln.db.select.dobject(dobject.id).one()
dobject
```

## Delete

Let us delete a file:

```python
dobject = ln.db.select.dobject(name="paradisi05_laminopathic_nuclei").one()
```

```python
ln.db.delete.dobject(dobject.id)
```

Check that the data is indeed gone from the database.

```python
ln.db.select.dobject(name="paradisi05_laminopathic_nuclei").all()
```
