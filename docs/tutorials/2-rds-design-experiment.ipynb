## Register experiment level metadata

Let's query for our project of interest:

```python
projects = ln.db.query.project(proj_id="P003")
projects
```

Next, we register two experiment types:

```python
exp_type1_id = ln.db.insert.experiment_type(efo_id="EFO:0030033", name="gRNA-seq")
```

```python
exp_type2_id = ln.db.insert.experiment_type(efo_id="EFO:0008860", name="Perturb-Seq")
```

Next, we register two experiments that will be performed in this project:

```python
experiment1_id = ln.db.insert.experiment(
    exp_id="EXP001",
    name=(
        "Genome-wide CRISPRi screen for functional regulators of cytokine production in"
        " response to stimulation"
    ),
    project_id=projects[0].id,
    experiment_type_id=exp_type1_id,
)
```

```python
experiment2_id = ln.db.insert.experiment(
    exp_id="EXP002",
    name=(
        "CRISPRa Perturb-seq for characterizing molecular phenotypes of cytokine"
        " regulators"
    ),
    project_id=projects[0].id,
    experiment_type_id=exp_type2_id,
)
```

## Register biosample level metadata

Next, we register a biosample that will be used in the experiments:

Along with some biological entities that will be used for annotating the biosample:

```python
species_id = ln.db.insert.species(common_name="human")
```

```python
cell_type_id = ln.db.insert.cell_type(ontology_id="CL:0000084", name="T cell")
```

```python
tissue_id = ln.db.insert.tissue(
    ontology_id="CL:2000001", name="peripheral blood mononuclear cell"
)
```

```python
biosample_id = ln.db.insert.biosample(
    sample_id="S1",
    name="primary human T cell",
    batch="1",
    species_id=species_id,
    cell_type_id=cell_type_id,
    tissue_id=tissue_id,
)
```

## Register readouts

Next, we register two types of readout that will be captured:

```python
readout1_id = ln.db.insert.readout(efo_id="EFO:0003024")
```

```python
readout2_id = ln.db.insert.readout(efo_id="EFO:0008913")
```

## Register featuresets

Here we also register two sets of features that will be measured in the experiments:

```python
featureset1_id = ln.db.insert.featureset(
    feature_entity="gene", name="CRISPRa-geneset-1"
)
```

```python
featureset2_id = ln.db.insert.featureset(
    feature_entity="gene", name="Perturbseq-geneset-1"
)
```

## Register biometa

biometa holds the following 4 fields and is directly link to datasets.

```python
biometa1_id = ln.db.insert.biometa(
    experiment_id=experiment1_id,
    biosample_id=biosample_id,
    readout_id=readout1_id,
    featureset_id=featureset1_id,
)
```

```python
biometa2_id = ln.db.insert.biometa(
    experiment_id=experiment2_id,
    biosample_id=biosample_id,
    readout_id=readout2_id,
    eatureset_id=featureset2_id,
)
```
